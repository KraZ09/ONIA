```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

# T1
task1 = test['text'].str.len()

# T2, \b means word boundary so it doesn t count words like 'freedom'
task2 = test['text'].str.count(r'\bfree\b')

# T3
tfidf = TfidfVectorizer(max_features=None, stop_words='english')
X_train = tfidf.fit_transform(train['text'])
X_test = tfidf.transform(test['text'])

model = LogisticRegression(max_iter=1000)
model.fit(X_train, train['label'])
task3 = model.predict_proba(X_test)[:, 1]  # probabilitatea clasei 1 (spam)

# Submission
rows = []
for sid, t1, t2, t3 in zip(test['sample_id'], task1, task2, task3):
    rows.append({'subtaskID': 1, 'datapointID': sid, 'answer': t1})
    rows.append({'subtaskID': 2, 'datapointID': sid, 'answer': t2})
    rows.append({'subtaskID': 3, 'datapointID': sid, 'answer': round(t3, 4)})

sub = pd.DataFrame(rows)
sub.to_csv('submission.csv', index=False)
```
